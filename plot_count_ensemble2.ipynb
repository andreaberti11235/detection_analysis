```python
import os
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

## Variazione soglia IOU (detection)

Variazione soglia, avendo fissato IOU per detection a 0.1. Analisi con ensemble ottenuto tramite media degli score di conf.

Total number of masses = 33

```python
txt_file = '/Users/andreaberti/Documents/work/DBT/test_inference/plot_ensemble/detections_mean_iou_1e-1.txt'
df_mean = pd.read_csv(txt_file, sep= ' ')
```

```python
df_mean
```

```python
fig, ax1 = plt.subplots()
ax1.set_xlabel('Conf threshold')
ax1.set_ylabel('Counts TP and FN')
ax1.plot(df_mean.loc[:, 'conf_th'], df_mean.loc[:, 'nr_TP'], '-o', label='True Positives')
ax1.plot(df_mean.loc[:, 'conf_th'], df_mean.loc[:, 'nr_FN'], '-o', label='False Negatives')

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:green'
ax2.set_ylabel('Counts FP', color=color)
ax2.plot(df_mean.loc[:, 'conf_th'], df_mean.loc[:, 'nr_FP'], '-o', color=color, label='False Positives')
ax2.tick_params(axis='y', labelcolor=color)

fig.legend()
#fig.tight_layout()
```

## Comparison with previous case

Confronto tra ensemble ottenuto facendo la media e ensemble ottenuto facendo la somma.

```python
txt_file = '/Users/andreaberti/Documents/work/DBT/test_inference/plot_ensemble/detections_mean_iou_1e-1.txt'
df_mean = pd.read_csv(txt_file, sep= ' ')

txt_file = '/Users/andreaberti/Documents/work/DBT/test_inference/plot_ensemble/detections_conf_originals.txt'
df_sum = pd.read_csv(txt_file, sep= ' ')
```

```python
df_sum
```

```python
iou_th = np.array([0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.40, 0.45, 0.50, 0.55, 0.60, 0.65])
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(30, 10))
ax1, ax3 = ax[0], ax[1]
# First plot
ax1.set_xlabel('Conf threshold')
ax1.set_ylabel('Counts')
ax1.plot(df_mean.loc[:, 'conf_th'], df_mean.loc[:, 'nr_TP'], '-o', label='True Positives')
ax1.plot(df_mean.loc[:, 'conf_th'], df_mean.loc[:, 'nr_FN'], '-o', label='False Negatives')

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:green'
ax2.set_ylabel('Counts FP', color=color)
ax2.plot(df_mean.loc[:, 'conf_th'], df_mean.loc[:, 'nr_FP'], '-o', color=color, label='False Positives')
ax2.tick_params(axis='y', labelcolor=color)

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

ax1.set_title('Ensemble with Mean, IOU_th=0.1')


# Second plot
ax3.set_xlabel('Conf threshold')
ax3.set_ylabel('Counts')
ax3.plot(iou_th, df_sum.loc[:, 'nr_TP'], '-o', label='True Positives')
ax3.plot(iou_th, df_sum.loc[:, 'nr_FN'], '-o', label='False Negatives')

ax4 = ax3.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:green'
ax4.set_ylabel('Counts FP', color=color)
ax4.plot(iou_th, df_sum.loc[:, 'nr_FP'], '-o', color=color, label='False Positives')
ax4.tick_params(axis='y', labelcolor=color)

ax3.legend(loc='upper left')
ax4.legend(loc='upper right')

ax3.set_title('Ensemble with SUM, IOU_th=0.1')

# Adjust the spacing between the subplots
plt.subplots_adjust(wspace=0.3)
```
